```python
n = 10
for i in range(0,n):   # recorro los números de 0 hasta 9
    if i%3 == 0:       # primera condición: si es divisible por 3 lo saltea
        continue
    if i == 7:       # segunda condición: si llega al 7 termina el bucle
        break
    try:               # me anticipo a los posibles errores haciendo la operación dentro de un try
       division = 100/i  # operación a realizar
       print('si i =', i, 'la división es =', 100/i) # imprimir en caso de tener un i óptimo y mostrar el resultado
    except ZeroDivisionError:                # primer posible error esperado (aunque no vamos a contar con el 0, lo filtra primer if)
       print('División por cero en i=0')     # se imprimiría este error en caso de existir el error al dividir por 0
    except ValueError:                       # segundo posible error esperado (aunque no contamos con negativos pues range(0,10))
       print('Índice negativo: i')           # se imprimiría este error en caso de ingresar un negativo
    else:                                    # si el try es true además de imprimir lo que indica en try se imprime lo que dice el else
       print('100/i =', division)
    finally:                                 # para cada itearción realizada se imprime 'Iteración i completada'para asegurar el flujo
       print('Iteración', i , 'completada')

n = 10
a = -1                                             # como quiero recorrer los impares y comenzar en el 1 debo sumar dos para el siguiente impar
while a < n:                                       # con el limite del while me aseguro el break en 7
    a += 2                                         # al sumar de a dos recorro los impares (1,3,5,7)
    if a%3 == 0:                                   # primera condición: si es divisible por 3 lo saltea
      continue
    try:
       division = 100/a                               # lo mismo que en el bucle anterior
       print('si a =', a, 'la división es =', 100/a)
    except ZeroDivisionError:
       print('División por cero en a=0')
    except ValueError:
       print('Índice negativo: a')
    else:
       print('100/a =', division)
    finally:
       print('Iteración', a ,'completada')
```
