##### (4) Escribir un programa que pregunte al usuario una cantidad a invertir, el interés anual y el número de años, y muestre por pantalla el capital obtenido en la inversión cada año que dura la inversión.

```python
capital_inicial = int(input("ingrese la cantidad a invertir"))
interes = int(input("ingrese el interés anual en porcentaje"))
años = int(input("ingrese los años"))
for año in range(1,años + 1):
    capital = capital_inicial*(1 + interes/100)**año
    capital = round(capital,3)
    print(f'Su capital en el año {año} será de {capital} pesos')
```

##### (3) Escribir un programa que almacene las asignaturas de un curso (por ejemplo Matemáticas, Física, Química, Historia y Lengua) en una lista, pregunte al usuario la nota que ha sacado en cada asignatura, y después las muestre por pantalla con el mensaje En < 'asignatura' > has sacado <'nota'> donde <'asignatura'> es cada una des las asignaturas de la lista y <'nota'> cada una de las correspondientes notas introducidas por el usuario.

```python
asignaturas = ["Matemáticas", "Física", "Química", "Historia", "Lengua"]
notas = []
for asign in asignaturas:
    calificacion = input(f"que nota te sacaste en {asign}?")
    notas.append(calificacion)
for index,asignatura in enumerate(asignaturas):
    print(f'En {asignatura} has sacado {notas[index]}')
```

##### (4) Escribir un programa que pregunte al usuario los números ganadores de la lotería primitiva, los almacene en una lista y los muestre por pantalla ordenados de menor a mayor.

```python
numeros_elegidos = input("seleccione 6 números separados por espacios")
numeros_elegidos = numeros_elegidos.split(" ")
lista_numeros = []
for numero in numeros_elegidos:
    numero = int(numero)
    lista_numeros.append(numero)
lista_numeros.sort()
print(lista_numeros)
```
