```python
def piramide (n):
    lista = [str(num)*(n-num) for num in range(n)]
    for digit in lista:
        print( " ".join(digit))

piramide(8)

```

```python
lista1 =[1,2,6,7,9,14]
lista2 = [2,6,9,11,16]
lista1_impares = [x for x in lista1 if x%2 == 1]
lista2_pares = [x for x in lista2 if x%2 == 0] 
lista_final = lista1_impares + lista2_pares
print(lista_final)
```
