## Cadenas (Strings)
#### 1. Pide al usuario una frase y muestra cuántas palabras tiene.

#### 2. Solicita un nombre y muestra sus iniciales en mayúsculas.

#### 3. Pide una palabra y verifica si es un palíndromo.

#### 4. Reemplaza todas las vocales de una frase por *

#### 5. Mezcla: pide una frase y cuenta cuántas veces aparece cada letra usando un diccionario.

## Condicionales
#### 1. Pide un número y determina si es positivo, negativo o cero.

#### 2. Solicita una nota (0–100) y muestra si es “Aprobado” o “Reprobado”.

#### 3. Pide un año y determina si es bisiesto.

#### 4. Pide dos números y muestra cuál es mayor o si son iguales.

#### 5. Mezcla: pide un número y una palabra; si el número es par, imprime la palabra en mayúsculas, si es impar, en minúsculas.

## Bucles
#### 1. Imprime los números del 1 al 20 con un for.

#### 2.Pide una palabra y muestra cada letra junto con su posición.

#### 3. Calcula la suma de los primeros 100 números naturales.

#### 4. Pide números al usuario hasta que escriba 0, luego muestra la suma total.

#### 5. Mezcla: pide una lista de nombres y usa un bucle para mostrar solo los que tengan más de 5 letras.

## Listas
#### 1. Crea una lista con 10 números y muestra el promedio.

#### 2. Pide 5 palabras y guárdalas en una lista, luego ordénalas alfabéticamente.

#### 3. Elimina los duplicados de una lista dada.

#### 4. Pide números y guarda solo los pares en una nueva lista.

#### 5. Mezcla: pide una lista de números y usa condicionales para separar pares e impares en dos listas distintas.

## Tuplas
#### 1. Crea una tupla con los meses del año y muestra el primero y el último.


#### 2. Convierte una lista de números en tupla y muestra su longitud.

#### 3. Pide 3 coordenadas (x, y) y guárdalas en una tupla.

#### 4. Crea una tupla con 5 colores y verifica si contiene “rojo”.


#### 5. Mezcla: pide una lista de números, conviértela en tupla y muestra el máximo y mínimo.

## Diccionarios
#### 1. Crea un diccionario con claves "nombre", "edad", "ciudad" y pide datos al usuario.



#### 2. Haz un diccionario de precios de productos y permite consultar el precio ingresando el nombre.

#### 3. Pide 3 palabras en inglés y sus traducciones al español, guárdalas en un diccionario y permite consultar.

#### 4. Crea un diccionario con nombres y notas, luego muestra el promedio.

#### 5. Mezcla: pide una frase y guarda en un diccionario la frecuencia de cada palabra.

## Funciones
#### 1. Escribe una función que reciba un número y devuelva su cuadrado.


#### 2. Crea una función que reciba una lista y devuelva solo los elementos únicos.


#### 3. Haz una función que reciba una palabra y diga si es palíndromo.

#### 4. Crea una función que reciba dos números y devuelva el mayor.

#### 5. Mezcla: función que reciba un diccionario de estudiantes y devuelva el nombre del que tiene la nota más alta.

## Excepciones
#### 1. Pide dos números y divide el primero por el segundo, manejando división por cero.


#### 2. Pide un número y convierte la entrada a entero, manejando errores si el usuario escribe texto.

#### 3. Pide un índice y muestra el elemento de una lista, manejando error si el índice no existe.

#### 4. Pide un archivo para abrir y maneja la excepción si no se encuentra.

#### 5. Mezcla: pide números hasta que el usuario escriba algo que no sea número, maneja la excepción y termina mostrando la suma.
