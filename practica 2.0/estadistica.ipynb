```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

n = 100

generos = ["M", "F"] 
paises = ["Argentina", "México", "Chile", "España"] 
categorias_libro = ["Novela", "Ciencia", "Infantil", "Historia", "Autoayuda"] 
metodos_pago = ["Tarjeta", "PayPal", "Transferencia"] 
membresia = ["Sí", "No"]

data = { "cliente_id": range(1, n+1), "genero": np.random.choice(generos, n), "edad": np.random.randint(18, 70, n), "pais": np.random.choice(paises, n), "categoria_libro": np.random.choice(categorias_libro, n), "precio_libro": np.round(np.random.uniform(5, 50, n), 2), "cantidad_comprada": np.random.randint(1, 6, n), "metodo_pago": np.random.choice(metodos_pago, n), "membresia": np.random.choice(membresia, n, p=[0.3, 0.7]),"reseña": np.random.randint(1, 6, n), "tiempo_envio_dias": np.random.randint(1, 16, n) }

df = pd.DataFrame(data)

df

```

```python
sns.histplot(df["edad"], bins=10, kde=True) 
plt.title("Distribución de edades de clientes") 
plt.show()
```

```python
sns.countplot(x="categoria_libro", data=df)
plt.title("Categorías de libros más vendidas") 
plt.show()
```

```python
sns.boxplot(x="categoria_libro", y="precio_libro", data=df)
plt.title("Distribución de precios por categoría") 
plt.show()
```

```python
sns.scatterplot(x="edad", y="cantidad_comprada", data=df)
plt.title("Edad vs cantidad de libros comprados")
plt.show()
```

```python
sns.countplot(x="pais", hue="metodo_pago", data=df)
plt.title("Método de pago más usado por país")
plt.show()
```

```python
data = { "edad": [1.00, -0.30, -0.45, -0.50], "gasto": [-0.30, 1.00, 0.65, 0.55], "visitas": [-0.45, 0.65, 1.00, 0.40],
         "satisfacción": [-0.50, 0.55, 0.40, 1.00]}
corr_matrix = pd.DataFrame(data, index=["edad","gasto","visitas","satisfacción"], columns=["edad","gasto","visitas","satisfacción"])
plt.figure(figsize=(6,4)) 
sns.heatmap(corr_matrix, annot=True, cmap="RdBu", center=0, fmt=".2f") 
plt.title("Matriz de correlación entre variables") 
plt.show()
```
