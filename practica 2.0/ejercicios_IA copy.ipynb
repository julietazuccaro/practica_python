## Cadenas (Strings)
#### 1. Pide al usuario una frase y muestra cuántas palabras tiene.

```python
frase = input('ingrese una frase')
cantidad_de_palabras = frase.count(' ')
print(cantidad_de_palabras + 1)
```

#### 2. Solicita un nombre y muestra sus iniciales en mayúsculas.

#### 3. Pide una palabra y verifica si es un palíndromo.

```python
palabra = input('ingrese una palabra')
palabra = palabra.lower() 
if palabra[:] == palabra[::-1]:
    print('Es un palíndromo')
else:
    print ('No es un palíndromo')
```

#### 4. Reemplaza todas las vocales de una frase por *

```python
frase = input('Ingrese una frase')
for vocal in 'aeiouAEIOU':
    nueva_frase = nueva_frase.replace(vocal, '*')
print (nueva_frase)
```

#### 5. Mezcla: pide una frase y cuenta cuántas veces aparece cada letra usando un diccionario.

```python
frase = input('ingrese una frase')
frase = frase.replace(' ','')

conteo = {}
for letra in frase:
    if letra in conteo: 
        conteo[letra] += 1
    else:
        conteo[letra] = 1
print(conteo)
```

## Condicionales
#### 1. Pide un número y determina si es positivo, negativo o cero.

```python
numero = int(input('ingrese un numero'))
if numero < 0:
    print('el número es negativo')
elif numero >0:
    print('el número es positivo')
else:
    print('el número es 0')
```

#### 2. Solicita una nota (0–100) y muestra si es “Aprobado” o “Reprobado”.

```python
nota = float(input('ingrese una nota'))
if nota >= 0 and nota <= 60:
    print('estás desaprobado')
elif  nota > 60 and nota <= 100:
    print('estás aprobado')
else:
    print ('no es una nota válida')
```

#### 3. Pide un año y determina si es bisiesto.

```python
año = int(input('ingrese el año'))
if año % 100 == 0:
    if año % 400 == 0:
        print('el año es bisiesto')
    else:
        print('el año no es bisiesto')
elif año % 4 == 0:
    print('el año es bisiesto')
else:
    print('el año no es bisiesto')
```

#### 4. Pide dos números y muestra cuál es mayor o si son iguales.

```python
numero1 = int(input('escribe un número'))
numero2 = int(input('escribe otro número'))
if numero1 > numero2:
    print('el número', numero1, 'es mayor')
elif numero2 > numero1:
    print('el número', numero2, 'es mayor')
else:
    print('los números son iguales')

```

#### 5. Mezcla: pide un número y una palabra; si el número es par, imprime la palabra en mayúsculas, si es impar, en minúsculas.

```python
numero = int(input('ingrese un numero'))
palabra = input('ingrese una palabra')
if numero%2 == 0:
    print(palabra.upper())
else:
    print(palabra.lower())
```

#### 2.Pide una palabra y muestra cada letra junto con su posición.

#### 3. Calcula la suma de los primeros 100 números naturales.

```python
n = 1
suma = 0
while n <101:
    suma = n + suma
    n +=1
print(suma)

```

#### 4. Pide números al usuario hasta que escriba 0, luego muestra la suma total.

```python
numero = int(input('ingrese un número'))
suma = 0
while numero != 0:
    suma = suma + numero
    numero = int(input('ingrese otro número'))
print(suma)
```

#### 5. Mezcla: pide una lista de nombres y usa un bucle para mostrar solo los que tengan más de 5 letras.

```python
lista_de_nombres=['paula', 'pedro', 'angelica', 'julieta', 'maria', 'penelope']
lista_final= []
for nombre in lista_de_nombres:
    if len(nombre)> 5:
        lista_final.append(nombre)
print(lista_final)
```
