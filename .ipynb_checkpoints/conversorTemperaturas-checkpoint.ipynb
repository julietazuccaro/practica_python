# Conversor de temperaturas

## Entrada

Para convertir una temperatura la cual el usuario elige, primero debemos pedirle que ingrese dicho valor mediante un `input` (entrada). Además, en la misma línea transformamos la cadena con `float` en un número flotante, osea, con decimales.

```python
temperatura_celsius = float(input("Por favor ingrese una temperatura"))
```

## Proceso

A continuación declaramos e inicializamos la variable `temperatura_fahrenheit` con su respectiva fórmula, la cual es: **F = C × 1.8 + 32**. Aquí se reemplaza F con `temperatura_fahrenheit` y C con `temperatura_celsius`.

```python
temperatura_fahrenheit = temperatura_celsius * 1.8 + 32
```

## Salida

Para finalizar imprimimos el resultado en pantalla, y usamos `round` para recortar la cantidad de decimales a 2 o menos.

```python
print(f"Su temperatura en grados fahrenheit es: {round(temperatura_fahrenheit, 2)}")
```

### A continuación el código completo y ejecutable.

````python
temperatura_celsius = float(input("Por favor ingrese una temperatura: "))

temperatura_fahrenheit = temperatura_celsius * 1.8 + 32

print(f"Su temperatura en grados fahrenheit es: {round(temperatura_fahrenheit, 2)}")
````
