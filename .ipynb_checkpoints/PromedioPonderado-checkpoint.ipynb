# Activada Dos - Promedio


Primero debemos crear la funcion _promedio_ponderado_, teniendo en cuenta los parametros  utilizando *args.

```python
# Creamos la funcion utilizando la palabra clave "def".

def promedio_ponderado(a,b,*args):
    # Documentamos 
    '''
    Esta funcion calcula el promedio ponderado. Teniendo en cuenta
    los argumentos. Se puede calcular a partir de dos valores en adelante.
    '''
    # Creamos una variable para guardar la suma
    # Tenemos que sumar los valores de a y b a la suma 
    suma = a + b
    
    for i in args:
        # Vamos acumulando el valor de i sumandolo.
        suma += i
    # Utilizamos el len para calcular el total de argumentos. El + 2 es por las variables por defecto a y b.
    
    total = len(args) + 2

    # Calcularemos el promedio.
    promedio = suma / total

    # Tenemos que retornar el valor del promedio.
    return promedio

# Para ver si funciona tenemos que llamar a la funcion colocando los argumentos.
# Utilizamos el print para mostralo en pantalla.

# Por ejemplo 2,10,7.
print(promedio_ponderado(2,10,7))
# Por ejemplo 2,8
print (promedio_ponderado(2,8))
# Llamando a la funcion para saber que hace.
help(promedio_ponderado)
```
