## Promedio Ponderado

##### Para calcular el promedio ponderado utilizaremos la fórmula : Promedio = (Suma de todos los números) / (Cantidad total de números)

```python
def promedio_poderado(n1,n2,*args):
    sumatoria = sum(args) + n1 + n2  
    cantidad_args = len(args) + 2
    promedio = sumatoria/cantidad_args
    return promedio
```

```python
promedio_poderado(1,5,6,7,9,12)
```
