```python
def promedio_ponderado(*args):
    """
    calcula el promedio de una cantidad arbitraria de n√∫meros.
    """
    if len(args)==0:
        return 0

        suma_total= sum(args)
        cantidad_elementos= len(args)
        resultado= suma_total/ cantidad_elementos

        return resultado
```

```python
resultado= promedio_ponderado( 10, 20, 30)
```

```python
print(f"El promedio es:{resultado}")
```
