# Conversor entre Celsius y Farenheit
## _Fórmula para convertir grados celcius en grados Farenheit_


**F= Cx1.8 + 32**

```python
# Se ingresará la temperatura en ºC que se quiere convertir
celcius = float( input("Ingrese la temperatura en celcius: "))
# El valor ingresado se convertirá utilizando la fórmula de conversión
farenheit = celcius*1.8 + 32
 
# Se da salida al valor obtenido
print(str(celcius)+"ºC equivalen a " + str(farenheit)+"ºF")
```
