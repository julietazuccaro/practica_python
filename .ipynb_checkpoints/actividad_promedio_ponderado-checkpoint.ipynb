## Creación de la función promedio_ponderado


Objetivo: Aplicar modularidad utilizando parámetros fijos y arbitrarios (*args), paréntesis, asterisco y la sentencia return. No se utiliza print dentro de la función.

```
def promedio_ponderado(divisor, *args):
    suma = sum(args)
    promedio = suma / divisor
    return promedio
```

### Ejemplo de uso

```
resultado = promedio_ponderado(3, 7, 8, 9)
resultado
```
